# Implement Gradient Descent for Linear Regression

## Implement the gradient descent algorithm for a simple linear regression problem.

###  $$ y=2*x+1$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = 2 * X + 1
```

```python
w = 0.0
b = 0.0
lr = 0.01
```

```python
for i in range(1000):
      y_pred = w * X + b
      dw = (-2/len(X)) * sum((y - y_pred) * X)
      db = (-2/len(X)) * sum(y - y_pred)
      w = w - lr * dw
      b = b - lr * db
```

```python
print("Final weight (w):", w)
print("Final bias (b):", b)
y_pred_final = w * X + b
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Original data')
plt.plot(X, y_pred_final, color='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression via Gradient Descent')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
