## Power Rule & Subplots Practice

Given function:  
**f(x) = 3x³ + 2x² − 5x + 7**


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(X):
  return 3*X**3+2*X**2-5*X+7
```

```python
def df(x):
    return 9 * x**2 + 4 * x - 5
```

```python
x_1 = np.linspace(-10, 10, 400)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('function')
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))

plt.subplot(1, 2, 2)
plt.title('derivative')
plt.xlabel('X', fontsize=16)
plt.ylabel("f'(x)", fontsize=16)
plt.plot(x_1, df(x_1))
plt.show()
```
