#  Learning Rate Effect

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 1

def df(x):
    return 2*x
```

```python
def gradient_descent(start_x, learning_rate, precision=0.0001, max_iters=1000):
    x = start_x
    x_vals = [x]

    for _ in range(max_iters):
        grad = df(x)
        new_x = x - learning_rate * grad
        step = abs(new_x - x)
        x = new_x
        x_vals.append(x)

        if step < precision:
            break

    return x_vals
```

```python
x_init = 5

x_vals_001 = gradient_descent(x_init, learning_rate=0.01)
x_vals_01  = gradient_descent(x_init, learning_rate=0.1)
x_vals_05  = gradient_descent(x_init, learning_rate=0.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals_001, label='Learning Rate = 0.01', linewidth=2)
plt.plot(x_vals_01,  label='Learning Rate = 0.1', linewidth=2)
plt.plot(x_vals_05,  label='Learning Rate = 0.5', linewidth=2)
plt.xlabel("Iterations", fontsize=14)
plt.ylabel("x values", fontsize=14)
plt.title("Effect of Learning Rate on Convergence", fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
